```python
import unittest
from recipes.Veg.TestVegNut import TestNutrition  #Test module for calories of Veg dishes(I/O is suppressed in this)
from recipes.NonVeg.TestNonVegNut import TestNonNutrition #Test module for calories of Non-Veg dishes(I/O is suppressed in this)
from recipes.Veg.TestVegEasy import TestVegEasy #Test module for easy vegetarian recipes
from recipes.NonVeg.TestNonVegEasy import TestNonVegEasy #Test module for easy non-vegetarian recipes
from recipes.Veg.TestVegMedium import TestVegMedium #Test module for medium vegetarian recipes
from recipes.NonVeg.TestNonVegMedium import TestNonVegMedium #Test module for medium non-vegetarian recipes
from recipes.Veg.TestVegDiff import TestVegDifficult #Test module for difficult vegetarian recipes
from recipes.NonVeg.TestNonVegDifficult import TestNonVegDifficult #Test module for difficult non-vegetarian recipes


def my_suite():
    suite = unittest.TestSuite()
    result = unittest.TestResult() 
    suite.addTest(unittest.makeSuite(TestNutrition)) 
    suite.addTest(unittest.makeSuite(TestNonNutrition))
    suite.addTest(unittest.makeSuite(TestVegEasy))
    suite.addTest(unittest.makeSuite(TestNonVegEasy))
    suite.addTest(unittest.makeSuite(TestVegMedium))
    suite.addTest(unittest.makeSuite(TestNonVegMedium))
    suite.addTest(unittest.makeSuite(TestVegDifficult))
    suite.addTest(unittest.makeSuite(TestNonVegDifficult))
    runner = unittest.TextTestRunner() 
    print(runner.run(suite))
my_suite()
```
